```python
import pandas as pd

def get_database():
  url = "https://raw.githubusercontent.com/SebastienBienfait/L2C-Data-managment/e1c8c253623aa878ce63b0d224269cfe1a4d4b20/Datasets/titanic_data.csv"
  data_table = pd.read_csv(url)
  #display(data_table.info())
  data_table.drop(data_table[data_table['Age'] == float("NaN")].index, inplace = True)
  return data_table, data_table["PassengerId"].tolist(),data_table["Pclass"].tolist(),data_table["Survived"].tolist(), data_table["Sex"].tolist(), data_table["Age"].tolist()



df, ID_list,class_list, survived_list, sex_list, age_list  = get_database()

#display(df)

#survival_rate = lambda x: sum(x)/len(x)
#total_survived = survival_rate(survived_list)



def into_dictionary(list1,list2,list3,list4):
  dictionary_ = []
  for i in range(len(list1)): #range(20):#

    #dictionary_[list1[i]] = list2[i]
    dictionary_.append((list1[i],list2[i],list3[i],list4[i]))
  return dictionary_



def survival_rate(tuple_list):
  total = 0
  dead  = 0
  for passenger in tuple_list:
    if passenger[3] == 1:
      total += 1
    else:
      total += 1
      dead += 1
  return total, dead


def total_pop_survival_rates():
  total_pop_dict = into_dictionary(class_list,sex_list,age_list,survived_list)
  pop_sorted_list = sorted(total_pop_dict) #, key= lambda x: x[2]

  for passenger in pop_sorted_list:
    if pd.isna(passenger[2]) == True:
      pop_sorted_list.remove(passenger)

  class1_male_under40 = [] #class - sex - age - survived
  class1_male_over40 = []  #  0      1     2       3

  class1_female_under40 = []
  class1_female_over40 = []

  class2_male_under40 = []
  class2_male_over40 = []

  class2_female_under40 = []
  class2_female_over40 = []

  class3_male_under40 = []
  class3_male_over40 = []

  class3_female_under40 = []
  class3_female_over40 = []

  for passenger in pop_sorted_list:
    if passenger[0] == 1:
      if passenger[1] == "male":
        if passenger[2] < 40.0:
          class1_male_under40.append(passenger)
        else:
          class1_male_over40.append(passenger)

      elif passenger[1] == "female":
        if passenger[2] < 40.0:
          class1_female_under40.append(passenger)
        else:
          class1_female_over40.append(passenger)
    elif passenger[0] == 2:
      if passenger[1] == "male":
        if passenger[2] < 40.0:
          class2_male_under40.append(passenger)
        else:
          class2_male_over40.append(passenger)
          
      elif passenger[1] == "female":
        if passenger[2] < 40.0:
          class2_female_under40.append(passenger)
        else:
          class2_female_over40.append(passenger)
    elif passenger[0] == 3:
      if passenger[1] == "male":
        if passenger[2] < 40.0:
          class3_male_under40.append(passenger)
        else:
          class3_male_over40.append(passenger)
          
      elif passenger[1] == "female":
        if passenger[2] < 40.0:
          class3_female_under40.append(passenger)
        else:
          class3_female_over40.append(passenger)
    
  class1_male_under40_pop, class1_male_under40_dead = survival_rate(class1_male_under40)
  class1_male_over40_pop, class1_male_over40_dead = survival_rate(class1_male_over40)

  class1_female_under40_pop, class1_female_under40_dead = survival_rate(class1_female_under40)
  class1_female_over40_pop, class1_female_over40_dead = survival_rate(class1_female_over40)

  class2_male_under40_pop, class2_male_under40_dead = survival_rate(class2_male_under40)
  class2_male_over40_pop, class2_male_over40_dead = survival_rate(class2_male_over40)
  
  class2_female_under40_pop, class2_female_under40_dead = survival_rate(class2_female_under40)
  class2_female_over40_pop, class2_female_over40_dead = survival_rate(class2_female_over40)

  class3_male_under40_pop, class3_male_under40_dead = survival_rate(class3_male_under40)
  class3_male_over40_pop, class3_male_over40_dead = survival_rate(class3_male_over40)
  
  class3_female_under40_pop, class3_female_under40_dead = survival_rate(class3_female_under40)
  class3_female_over40_pop, class3_female_over40_dead = survival_rate(class3_female_over40)

  print("death rate of 1st class men over 40: ", round((class1_male_over40_dead*100)/class1_male_over40_pop,2),"%")
  print("death rate of 1st class women over 40: ", round((class1_female_over40_dead*100)/class1_female_over40_pop,2),"%")
  print("death rate of 1st class men under 40: ", round((class1_male_under40_dead*100)/class1_male_under40_pop,2),"%")
  print("death rate of 1st class women under 40: ", round((class1_female_under40_dead*100)/class1_female_under40_pop,2),"%")

  print("death rate of 2nd class men over 40: ", round((class2_male_over40_dead*100)/class2_male_over40_pop,2),"%")
  print("death rate of 2nd class women over 40: ", round((class2_female_over40_dead*100)/class2_female_over40_pop,2),"%")
  print("death rate of 2nd class men under 40: ", round((class2_male_under40_dead*100)/class2_male_under40_pop,2),"%")
  print("death rate of 2nd class women under 40: ", round((class2_female_under40_dead*100)/class2_female_under40_pop,2),"%")

  print("death rate of 3rd class men over 40: ", round((class3_male_over40_dead*100)/class3_male_over40_pop,2),"%")
  print("death rate of 3rd class women over 40: ", round((class3_female_over40_dead*100)/class3_female_over40_pop,2),"%")
  print("death rate of 3rd class men under 40: ", round((class3_male_under40_dead*100)/class3_male_under40_pop,2),"%")
  print("death rate of 3rd class women under 40: ", round((class3_female_under40_dead*100)/class3_female_under40_pop,2),"%")

  request = ["","",""]
  request[0] = input("What class ticket of passenger: ")
  request[1] = input("What gender (male/female): ")
  request[2] = input("What age group (under/over 40): ")

  if request[2] == "over ":
    if request[1] == "male":
      if request[0] == "1":
        print("survival rate of 1st class men over 40: ", round(((class1_male_over40_pop - class1_male_over40_dead)*100)/class1_male_over40_pop,2),"%")
      elif request[0] == "2":
        print("survival rate of 2nd class men over 40: ", round(((class2_male_over40_pop - class2_male_over40_dead)*100)/class2_male_over40_pop,2),"%")
      else:
        print("survival rate of 3rd class men over 40: ", round(((class3_male_over40_pop - class3_male_over40_dead)*100)/class3_male_over40_pop,2),"%")
    else:
      if request[0] == "1":
        print("survival rate of 1st class women over 40: ", round(((class1_female_over40_pop - class1_female_over40_dead)*100)/class1_female_over40_pop,2),"%")
      elif request[0] == "2":
        print("survival rate of 2nd class women over 40: ", round(((class2_female_over40_pop - class2_female_over40_dead)*100)/class2_female_over40_pop,2),"%")
      else:
        print("survival rate of 3rd class women over 40: ", round(((class3_female_over40_pop - class3_female_over40_dead)*100)/class3_female_over40_pop,2),"%")
  else:
    if request[1] == "male":
      if request[0] == "1":
        print("survival rate of 1st class men under 40: ", round(((class1_male_under40_pop - class1_male_under40_dead)*100)/class1_male_under40_pop,2),"%")
      elif request[0] == "2":
        print("survival rate of 2nd class men under 40: ", round(((class2_male_under40_pop - class2_male_under40_dead)*100)/class2_male_under40_pop,2),"%")
      else:
        print("survival rate of 3rd class men under 40: ", round(((class3_male_under40_pop - class3_male_under40_dead)*100)/class3_male_under40_pop,2),"%")
    else:
      if request[0] == "1":
        print("survival rate of 1st class women under 40: ", round(((class1_female_under40_pop - class1_female_under40_dead)*100)/class1_female_under40_pop,2),"%")
      elif request[0] == "2":
        print("survival rate of 2nd class women under 40: ", round(((class2_female_under40_pop - class2_female_under40_dead)*100)/class2_female_under40_pop,2),"%")
      else:
        print("survival rate of 3rd class women under 40: ", round(((class3_female_under40_pop - class3_female_under40_dead)*100)/class3_female_under40_pop,2),"%")
total_pop_survival_rates()
```
