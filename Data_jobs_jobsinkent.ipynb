```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

base_url = "https://jobsinkent.com/search?q=Data&pl=1" #&start=0

def get_html(url):
    #agent = {"User-Agent":"Mozilla/5.0"}
    page = requests.get(url) #, headers = agent

    soup = BeautifulSoup(page.content, 'html.parser') # 
    return soup

def clean_bloat(text):
  strip1 = text.strip("\n")
  strip2 = strip1.strip(" ")
  strip3 = strip2.strip("\n")
  return strip3

def find_jobs_iterate(url):
  jobsinkent_df = pd.DataFrame(columns = ["job_title", "company", "salary", "contract", "location"])
  site_html = get_html(url)
  results = site_html.find(class_="mt-2 lg:mt-0")

  job_cards = results.find_all(class_="flex mb-2 border-black-900 rounded-md border-x-4 sm:rounded-lg")

  for job in job_cards:
    job_title = job.find(class_="lg:text-2xl text-xl leading-6 font-bold hover:underline-offset-4").text
    company_name = job.find(class_="mt-0 mb-2 max-w-2xl text-sm text-black").text
    info_list = job.find_all(class_="col-span-12")# truncate mt-1 max-w-2xl text-sm text-black flex items-center")
    salary = clean_bloat(info_list[0].text)
    contract = clean_bloat(info_list[1].text)
    location = clean_bloat(info_list[2].text)


    new_row = {"job_title":job_title, "company":company_name, "salary":salary, "contract":contract, "location":location}

    jobsinkent_df = jobsinkent_df.append(new_row,ignore_index = True)

    print("\n")
  #print(job_cards.prettify())

  #return id_list
  return jobsinkent_df

jobsinkent_df = find_jobs_iterate(base_url)
```
